```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array(['preethi','priya','pooja','anu','sandya'])
y=np.array([30,65,50,40,80])
plt.title("college dataframe")
plt.xlabel("students")
plt.ylabel("percentage")
plt.bar(x,y, width=0.5)
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array(['cricket','vollyball','hocky','none'])
y=np.array([30,20,25,25])
plt.title("Sports Dataframe")
plt.xlabel("Games")
plt.ylabel("number of students")
plt.plot(x,y,marker='o',mec='r')
plt.show()
```
